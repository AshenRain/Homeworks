#### 2.8. Learning curve
Your goal is to estimate, how does the model behaviour change with the increase of the `train` dataset size.

* Split the training data into 10 equal (almost) parts. Then train the models from above (Logistic regression, Desicion Tree, Random Forest) with optimal hyperparameters you have selected on 1 part, 2 parts (combined, so the train size in increased by 2 times), 3 parts and so on.

* Build a plot of `accuracy` and `f1` scores on `test` part, varying the `train` dataset size (so the axes will be score - dataset size.

* Analyse the final plot. Can you make any conlusions using it?

```python
# YOUR CODE HERE
train_sizes = [len(X_train)//10*i for i in range(1, 11)]

models = [
    ('Logistic Regression', LogisticRegression(multi_class='multinomial', solver='saga', tol=1e-3, max_iter=100, C = 10, penalty = 'l1')),
    ('Decision Tree', DecisionTreeClassifier(criterion = 'entropy', max_depth=5)),
    ('Random Forest', RandomForestClassifier(max_depth=10, n_estimators=70))
]

for model_name, model in models:
    test_f1_scores = []
    test_accuracy_scores = []

    for train_size in train_sizes:
        # Обучаем модель на текущем размере обучающего набора
        model.fit(X_train[:train_size], y_train[:train_size])

        # Вычисляем предсказания на тестовом наборе
        test_predictions = model.predict(X_test)

        # Вычисляем f1-меру и точность на тестовом наборе
        test_f1 = f1_score(y_test, test_predictions, average = 'weighted')
        test_accuracy = accuracy_score(y_test, test_predictions)

        # Добавляем значения f1-меры и точности в списки
        test_f1_scores.append(test_f1)
        test_accuracy_scores.append(test_accuracy)

    # Строим графики зависимости точности и f1-меры от размера обучающего набора
    plt.plot(train_sizes, test_accuracy_scores, label='accuracy')
    plt.plot(train_sizes, test_f1_scores, label='f1')
    plt.xlabel('Размер обучающего набора')
    plt.ylabel('Значение метрики')
    plt.title(f'Кривая обучения для {model_name}')
    plt.legend()
    plt.show()
```

<font color="green"><b>
Explanation:
</b></font>


По графикам можно сделать заключения, что увеличения обучающей выборки не всегда идет к увеличению точности. Это можно объяснить тем, что когда мы добовляем следующую часть, в ней может содержаться большое кол-во шумов или пограничных ответов, которые могут спутать модель.
